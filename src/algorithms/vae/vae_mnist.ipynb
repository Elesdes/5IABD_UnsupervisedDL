```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
from keras.optimizers import Adam
from vae import VAE, VAE_Base

seed = 42
random.seed(seed)
np.random.seed(seed)
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = np.expand_dims(x_train, -1).astype("float32") / 255
x_test = np.expand_dims(x_test, -1).astype("float32") / 255
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
vae = VAE(input_dim=x_train.shape[1:], encoding_dim=2)

vae.compile(optimizer=Adam())

vae.fit(x_train, epochs=100, batch_size=512, validation_data=(x_test, None))
```

```python
vae.visualize_2d(x_train, y_train)
```

```python
vae.visualize_reconstruction(x_test)
```

```python
vae.generate_images()
```
