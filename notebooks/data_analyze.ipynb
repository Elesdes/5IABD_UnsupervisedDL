# Data Analyze

## MNIST Dataset

MNIST is a dataset of 60,000 28x28 grayscale images of the 10 digits, along with a test set of 10,000 images. It is a subset of a larger set available from NIST. The digits have been size-normalized and centered in a fixed-size image. 

## Load Dataset

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### Data Type

```python
print(f"{type(x_train) = }")
print(f"{type(x_test) = }")
print(f"{type(y_train) = }")
print(f"{type(y_test) = }")
```

### Shape

```python
print(f"{x_train.shape = }")
print(f"{x_test.shape = }")
print(f"{y_train.shape = }")
print(f"{y_test.shape = }")
```

### Visual Representation

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(3, 3, figsize=(8, 8))
indices = np.random.choice(range(x_train.shape[0]), size=9, replace=False)
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[indices[i]], cmap="gray")
    ax.set_title(f"Label: {y_train[indices[i]]}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```

### Data Distribution

```python
import pandas as pd

labels, counts = np.unique(y_train, return_counts=True)
df = pd.DataFrame({'Label': labels, 'Count': counts})
plt.figure(figsize=(10, 6))
plt.bar(df['Label'], df['Count'], color='skyblue')
plt.xlabel('Digit Label')
plt.ylabel('Count')
plt.title('Data Repartition of MNIST Dataset')
plt.show()
```
